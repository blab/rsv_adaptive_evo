```python
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
```

```python
#find where the placeholders for the duplication should be inserted
for record in SeqIO.parse(open('../rsv_testing-ref/config/rsv_a_on1_reference.gb', 'r'), "gb"):
    for feature in record.features:
        if feature.type =='CDS':
            if feature.qualifiers['gene'] == ['G']:
                g_seq = feature.location.extract(record.seq)
                g_loc_template = g_seq.find('GTCAAGAGGAAACCCTCCACTCAACCACCTCCGAAGGCTATCTAAGCCCATCACAAGTCTATACAACATCCG')
                loc_template = g_loc_template + feature.location.start
```

```python
aligned_other_file = '../rsv_testing-ref/results/aligned_A_other.fasta'

with open(aligned_other_file, 'r') as handle:
    records_with_placeholder = []
    
    for virus in SeqIO.parse(handle, 'fasta'):
#         template = virus.seq[loc_template-1:loc_template+71]
        virus.seq = virus.seq.tomutable()
        seq_with_placeholder = virus.seq[0:loc_template+71]+'-'*72+virus.seq[loc_template+71:]
        
        records_with_placeholder.append(SeqRecord(seq_with_placeholder, id=virus.id, description=virus.description))
    SeqIO.write(records_with_placeholder, '../rsv_testing-ref/data/rsv_A_other_withplaceholder.fasta', "fasta")
```

```python
#merge on1 and other RSV-A alignment files
rsv_a_other_withplaceholder_file = '../rsv_testing-ref/data/rsv_A_other_withplaceholder.fasta'
rsv_a_on1_file = '../rsv_testing-ref/results/aligned_A_on1.fasta'

all_a_records = []
with open(rsv_a_other_withplaceholder_file, 'r') as handle1:
    
    for virus in SeqIO.parse(handle1, 'fasta'):
        all_a_records.append(SeqRecord(virus.seq, id=virus.id, description=virus.description))
        
with open(rsv_a_on1_file, 'r') as handle2:
    
    for virus in SeqIO.parse(handle2, 'fasta'):
        all_a_records.append(SeqRecord(virus.seq, id=virus.id, description=virus.description))
```

```python
with open('../rsv_testing-merge/data/rsv_A_all_withplaceholder.fasta', 'w') as outfile:
    SeqIO.write(all_a_records, outfile, "fasta")
```
